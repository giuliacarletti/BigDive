```python
vectorizer = text.CountVectorizer(max_df=0.8,
                                  max_features=10000, 
                                  stop_words=text.ENGLISH_STOP_WORDS,
                                 ngram_range=(1,2))

                                # ngram_range mi dÃ  le coppie di parole 
  
    
counts = vectorizer.fit_transform(subreddits)  
tfidf = text.TfidfTransformer().fit_transform(counts)
```

```python
array(vectorizer.get_feature_names())[counts[0].nonzero()[1]] 
```

```python
from sklearn import decomposition

# Fit the NMF model
nmf = decomposition.NMF(n_components=6)   # NMF = Non Negative Matrix Factorization
nmf.fit(tfidf)
W = nmf.transform(tfidf)   # matrice W = peso
H = nmf.components_
```

```python
W.shape
```

```python
H.shape
```

```python
H[0].argsort()  # mi dÃ  gli indici ordinati per peso
# l'ultima parola ha il peso maggiore
```

```python
# Inverse the vectorizer vocabulary to be able
feature_names = vectorizer.get_feature_names()
```

```python
for topic_idx, topic in enumerate(H):
    print "Topic #%d:" % topic_idx
    print ",".join([feature_names[i] for i in topic.argsort()[:-21:-1]])
    print
```

```python
from scipy import sparse
```

```python
def extract_features(group):   # estrae le features di un dato gruppo
    group_subreddits = group['subreddit'].values     # del gruppo prendo tutti i subreddit
    idxs = subreddits_map[group_subreddits].values #uso la mappa di prima per mappare i subreddit da string a interi
    v = sparse.dok_matrix((1, subreddits.shape[0]))
    for idx in idxs:
        if not np.isnan(idx):
            v[0, idx] = +1 # provare anche con la multinominal naive based
    return v.tocsr()

extract_features(train_data[train_data.author=='RedThunder90'])
# ho una matrice di 1 x 3866 (cioe quell'autore ha fatto commenti solo per quel subreddit li = 1 stored element)
```

```python
features_dict = {}  # per quell'autore assegno le features = cioÃ¨ i commenti che lui ha fatto nei suoi subreddit

for author, group in train_data.groupby('author'):
    features_dict[author] = extract_features(group)   #ho un dizionario con per ogni autore il vettore corrisp
```

# Model Selection

```python
# YOUR CODE HERE

from sklearn import grid_search
from sklearn import metrics, cross_validation


from sklearn import linear_model
model = linear_model.LogisticRegression(penalty = 'l2')

Cs = np.logspace(-4, 0, 20)
gs = grid_search.GridSearchCV(model, {'C': Cs}, cv=10)
gs.fit(X, y)


print "best alpha:", gs.best_params_
```

# Prepare the solution

```python
test_data = pd.read_csv("test_data.csv.gz", compression="gzip", encoding="utf8")
```

```python
features_dict = {}

for author, group in test_data.groupby('author'):
    features_dict[author] = extract_features(group)
```

```python
X_test = sparse.vstack([features_dict[author] for author in test_data.author.unique()])
X_test
```
