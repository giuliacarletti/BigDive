Prepare a sinthetic dataset for this example.

```python
from sklearn import datasets

X, _ = datasets.make_blobs(n_samples=150, random_state=8, centers=4)
# X, _ = datasets.make_circles(n_samples=150, factor=.5, noise=.05)

plot(X[:, 0], X[:, 1], 'bx')
```

# Kmeans clustering

```python
np.random.seed(42)

from sklearn import cluster

kmeans = cluster.KMeans(n_clusters=3, init='random',  # init= algoritmo in cui i punti del centro son inizializzati
                        n_init=1, max_iter=10)
C = kmeans.fit_predict(X)

colors = array(['b', 'r', 'g', 'k'])   #  il range di colori
scatter(X[:, 0], X[:, 1], marker="x", c=colors[C])

centers = kmeans.cluster_centers_  # centroidi dei clusters  # i centroidi iniziali sono determinati a caso (random)
scatter(centers[:, 0], centers[:, 1], s=100, c=colors)
```

```python
# il grafico risultante dipende da come scelgo il punto iniziale
# cambio lo stato iniziale

np.random.seed(43)

from sklearn import cluster

kmeans = cluster.KMeans(n_clusters=4, init='random',  # init= algoritmo in cui i punti del centro son inizializzati
                        n_init=1, max_iter=10)    # metto 4 centri anziche 3
C = kmeans.fit_predict(X)

colors = array(['b', 'r', 'g', 'k'])   #  il range di colori
scatter(X[:, 0], X[:, 1], marker="x", c=colors[C])

centers = kmeans.cluster_centers_  # centroidi dei clusters  # i centroidi iniziali sono determinati a caso (random)
scatter(centers[:, 0], centers[:, 1], s=100, c=colors)
```

```python
# il grafico risultante dipende da come scelgo il punto iniziale
# cambio lo stato iniziale

np.random.seed(43)

from sklearn import cluster

kmeans = cluster.KMeans(n_clusters=5, init='random',  # init= algoritmo in cui i punti del centro son inizializzati
                        n_init=1, max_iter=10)    # metto 5 centri (colori) anziche 3
C = kmeans.fit_predict(X)

colors = array(['b', 'r', 'g', 'k', 'y'])   #  il range di colori
scatter(X[:, 0], X[:, 1], marker="x", c=colors[C])

centers = kmeans.cluster_centers_  # centroidi dei clusters  # i centroidi iniziali sono determinati a caso (random)
scatter(centers[:, 0], centers[:, 1], s=100, c=colors)
```

# Hierarchical clustering

```python
X, _ = datasets.make_blobs(n_samples=50, random_state=8, centers=4)
# X, _ = datasets.make_circles(n_samples=200, factor=.1, noise=.05)

scatter(X[:, 0], X[:, 1]);   # come dataset qui ho 4 nuovole, un po' piu sparse di prima
```

```python
from scipy.cluster.hierarchy import linkage, dendrogram, fcluster

Z = linkage(X, method='average', metric='euclidean')  # per ogni agglomerazione calcolo il punto medio (average)
figure(figsize=(15,5))
P = dendrogram(Z);
```

```python
T = fcluster(Z, 4, 'maxclust')   # la funzione fcluster taglia l'albero in qualche punto
                                    # qui non c'e nessun fattore di inizializzazione
figure(figsize=(5,5))
colors = array(['b', 'r', 'g', 'k'])
scatter(X[:, 0], X[:, 1], color=colors[T-1]);
```

### Hierarchical Clustering in Scikit-Learn:

```python
from sklearn import cluster

model = cluster.AgglomerativeClustering(n_clusters=4, affinity='euclidean', linkage='average')

clusters = model.fit_predict(X)

figure(figsize=(5,5))
colors = array(['b', 'r', 'g', 'k'])
scatter(X[:, 0], X[:, 1], color=colors[clusters-1]);
```

```python
from sklearn import cluster

model = cluster.AgglomerativeClustering(n_clusters=3, affinity='euclidean', linkage='average')

clusters = model.fit_predict(X)

figure(figsize=(5,5))
colors = array(['b', 'r', 'g', 'k'])
scatter(X[:, 0], X[:, 1], color=colors[clusters-1]);
```
